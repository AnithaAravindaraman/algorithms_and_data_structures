  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
# Define the problem: We are given a list of n integers from 0 to n. Find and return the missing integers within that range, if none missing, return the value '-1'
# My task is to find and return the missing numbers (if any), otherwise return -1

from typing import List

def missing_num(nums: List[int]) -> List[int]:
    n = max(nums) # Find the maximum number in the list
    full_range = set(range(n + 1)) # Create a set of all numbers from 0 to n
    missing_numbers = full_range - set(nums) # Find the difference between the full range and the numbers in the list
    return sorted(list(missing_numbers)) if missing_numbers else -1 #return list of sorted missing numbers or -1 if none are missing

print(missing_num([3, 0, 1]))         
print(missing_num([0, 1]))            
print(missing_num([9,6,4,2,3,5,7,0,1]))
print(missing_num([0, 1, 2, 3, 4, 4, 8, 9])) 
```


-   Copy the solution your partner wrote. 


```python
# My partner: https://github.com/c5ho/algorithms_and_data_structures/blob/main/02_activities/assignments/assignment_1.ipynb
from typing import List

def first_duplicate(nums: List[int]) -> int:
    
    # Create an empty set to track seen numbers
    seen_nums = set()

    # Iterate through the list of given numbers
    for num in nums:
        
        # If the number has been is in the seen list, return it
        if num in seen_nums:
            return num
        
        # Otherwise, add the number to the seen list
        seen_nums.add(num)

    # If no duplicates are found, return -1
    return -1 
```


-   Explain why their solution works in your own words.


```python
# The solution works because it keeps track of numbers it's already seen using a set. As it goes through the list, it checks each number:

# If the number is already in the set, that means it's a duplicate, so it returns that number right away.

# If not, it adds the number to the set and keeps going.

#This way, the function finds the first number that repeats and returns it. If there are no duplicates, it just returns -1. Using a set makes the checking quick and easy.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# The time complexity is O(n) because the function looks at each number in the list one time. It stops early if it finds a duplicate, but in the worst case it checks every number once.

# The space complexity is also O(n) because it uses a set to keep track of all the numbers it has seen. In the worst case, if there are no duplicates, it stores every number in the set.
```
